# **2.Demostraciones de Periodicidad en Señales Discretas**
::

### **Señal a: $Acos(\omega_0n +\phi ) $**
Tenemos que igualar nuestra
$x(n)=Acos(\omega_0n +\phi )$ a x(n+N)
Lo cual nos quedaria de la siguiente manera:

$x(n+N)= Acos(\omega_0(n+N) + \phi)$

Para que sea periódica se tiene que cumplir que :
$\omega_0 N = 2\pi k, \quad k \in \mathbb{Z}$
Por lo tanto:
$\frac{\omega_0}{2\pi} =\frac{N}{K}$

Asi que si **Es periodica**.


### **Señal b : $x(n) = Ae^{j \omega_0 n}$**

Repitimos el mismo proceso igualamos a x(n+N)

$x(n+N) = A e^{j \omega_0 (n+N)}$

Por lo tanto:

$\frac{\omega_0}{2\pi} =\frac{N}{K}$

Asi que si **Es periodica**.


### **Señal c :$x(n) = e^{-j0.4n}$**

La igualamos a x(n+N)

$x(n + N) = A e^{jω₀(n + N)} = A e^{jω₀n} e^{jω₀N}$

Para que $x(n + N) = x(n)$, se requiere:

$e^{jω₀N} = 1 ⇒ ω₀N = 2πk ⇒ ω₀/(2π) = k/N$

Razón:
$\frac{\omega_0}{2\pi} = \frac{-0.4}{2\pi} \approx -0.06366$

Como el resultado no es entero entonces la señal **No es Periodica**


### **Señal d : $x(n) = 2\cos(1.1\pi n - 0.5\pi) + 2\sin(0.7\pi n)$**

Analizamos cada componente por separado:

Primera componente:

$x_1(n) = 2 cos(1.1\pi n - 0.5\pi )$

La frecuencia anguar es $\omega_1 = 1.1 \pi $

$\frac{\omega_1}{2\pi}$ = $\frac{1.1\pi }{2\pi}$ = $\frac{11}{20}(2\pi)$

$N=20$

Segunda Componente:
$x_2(n) = 2 cos(0.7\pi )$

La frecuencia anguar es $\omega_2 = 0.7 \pi $

$\frac{\omega_2}{2\pi}$ = $\frac{0.7\pi }{2\pi}$ = $\frac{7}{20}(2\pi)$

$N=20$

Como el resultado de ambas partes es entero la señal **Si es Periodica**.

### **Señal e:**
$x(n) = \cos(3n)$

Razón:
$\frac{\omega_0}{2\pi} = \frac{3}{2\pi} \approx 0.477$

Como la razón no es un entero , la señal **No es Periódica**.

##Solución del Problema 2

Condición para periodicidad en tiempo discreto

Una señal discreta $x(n) = \cos(\omega_0 n)$ es periódica si existe un entero $N > 0$ tal que:
$\cos(\omega_0 (n + N)) = \cos(\omega_0 n) \quad \text{para todo } n $

Esto requiere que:
$\omega_0 N = 2\pi k \quad \text{para algún entero } k$

Es decir:$
\frac{\omega_0}{2\pi} = \frac{k}{N} \quad \text{debe ser un número entero}
$

Dado que $x(n) = x_a(nT_s) = \cos(\Omega_0 nT_s)$, tenemos:

$\omega_0 = \Omega_0 T_s4$

Por lo tanto, la condición de periodicidad es:

$\frac{\Omega_0 T_s}{2\pi} = \frac{k}{N} \quad $


$x(n)$ es periódica si y solo si:

$\frac{\Omega_0 T_s}{2\pi} = \frac{k}{N} \quad \text{es un número racional}$

Es decir, si $\frac{\Omega_0 T_s}{2\pi}$ es racional.

##Caso específico con $\Omega_0 = 18 \, \text{rad}$ y $T_s = \pi/6 \, \text{s}$

Calculamos:
$\omega_0 = \Omega_0 T_s = 18 \times \frac{\pi}{6} = 3\pi $

Ahora verificamos la condición:
$\frac{\omega_0}{2\pi} = \frac{3\pi}{2\pi} = \frac{3}{2}$

Como $\frac{3}{2}$ es racional, la señal **Si es Peridica**.


El periodo fundamental $N$ es el menor entero positivo tal que:
$\omega_0 N = 2\pi k \quad \text{para algún entero } k$

Sustituyendo:

$3\pi N = 2\pi k \implies 3N = 2k \implies N = \frac{2k}{3}$

Buscamos el menor entero $N > 0$ tal que $\frac{2k}{3}$ sea entero. Esto ocurre cuando $k = 3$:
$N = \frac{2 \times 3}{3} = 2$

Por lo tanto, el periodo fundamental es $N = 2$.



# **3.Conclusión**
La periodicidad se refiere a la propiedad de una señal o fenómeno de repetirse en intervalos regulares de tiempo o espacio. Este concepto es fundamental en diversas áreas, como física, matemáticas e ingeniería, porque permite predecir el comportamiento de sistemas complejos a partir de un patrón repetitivo.Para determinar si una señal es periódica, podemos igualar nuestra señal $x(n)$ con $x(n+N)$ y, a partir de esta igualdad, obtener la relación $\frac{N}{K}$ Esta relación debe dar como resultado un número entero para confirmar que la señal es periódica. Además, gracias a las gráficas, podemos verificar cuáles señales son periódicas y cuáles no: en las periódicas, los valores se repiten regularmente, mientras que en las no periódicas, como las señales C y E, ningún valor se repite, aunque sus gráficas puedan parecer similares.
